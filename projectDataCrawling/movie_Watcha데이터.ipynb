```python
import time
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
# 초기화 버튼
watcha_movie_title_list = [] # 제목
watcha_movie_year_list = []  # 년도
watcha_movie_syn_list = []   # 시놉시스
watcha_movie_grade_list=[]   # 평점
watcha_movie_detail_list=[]  # 장르
```

```python
url = "https://www.justwatch.com/kr/%EB%8F%99%EC%98%81%EC%83%81%EC%84%9C%EB%B9%84%EC%8A%A4/netflix"
driver = wb.Chrome()
driver.get(url)
```

```python
driver.find_element(By.CSS_SELECTOR, "div.title-info.visible-xs.visible-sm > div:nth-child(3) > .detail-infos__value").text
```

```python
# 평점
# temp_list = []
# temp = driver.find_elements(By.CSS_SELECTOR, "div.jw-scoring-listing__rating")
# temp_list.append(temp[3].text)
```

```python
for i in range(50):
    driver.find_element(By.CSS_SELECTOR,"body").send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5)
```

```python
for i in range(51,52):
    driver.find_element(By.CSS_SELECTOR, f".title-list-grid > div:nth-child({i}) > a").click()
    time.sleep(6)
    title = driver.find_element(By.CSS_SELECTOR, ".title-block div h1").text
    time.sleep(7)
    year = driver.find_element(By.CSS_SELECTOR, ".title-block span").text
    time.sleep(6)
    syns = driver.find_element(By.CSS_SELECTOR,".text-wrap-pre-line.mt-0 > span").text
    time.sleep(6)
    temp = driver.find_elements(By.CSS_SELECTOR, "div.jw-scoring-listing__rating")
    defail = driver.find_element(By.CSS_SELECTOR, "div.title-info.visible-xs.visible-sm > div:nth-child(3) > .detail-infos__value").text
    watcha_movie_title_list.append(title)
    watcha_movie_year_list.append(year)
    watcha_movie_syn_list.append(syns)
    watcha_movie_grade_list.append(temp[3].text)
    watcha_movie_detail_list.append(defail)
    driver.back()
    if(i%10==0):
        for j in range(1,3):
            time.sleep(j)
    time.sleep(6)
```

```python
dic = {
    "제목":watcha_movie_title_list,
    "장르":watcha_movie_detail_list,
    "년도":watcha_movie_year_list,
    "줄거리": watcha_movie_syn_list,
    "평점":watcha_movie_grade_list
    }
```
