```python
import time
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd


```

```python
data = pd.read_csv("./REVIEW_DATA.csv", index_col=False)
data
```

```python
title = data["제목"]
len(title)
```

```python
url = "https://movie.naver.com/"
driver = wb.Chrome()
driver.get(url)
```

```python
movie_list = []
movie_title_list = []
movie_text_list=[]
```

```python
driver.find_element(By.CSS_SELECTOR,"div.score_result span.ico_viewer").text
```

```python

driver.find_element(By.CSS_SELECTOR,"#ipt_tx_srch").send_keys(title[1]+"\n")
driver.find_element(By.CSS_SELECTOR,"#old_content > ul.search_list_1 > li:nth-child(1) > dl > dt > a").click()
driver.find_element(By.CSS_SELECTOR,"#movieEndTabMenu > li:nth-child(5) > a").click()


```

```python
for i in range(28,len(title)):
    driver.find_element(By.CSS_SELECTOR,"#ipt_tx_srch").send_keys(title[i]+"\n")
    driver.find_element(By.CSS_SELECTOR,"#old_content > ul.search_list_1 > li:nth-child(1) > dl > dt > a").click()
    driver.find_element(By.CSS_SELECTOR,"#movieEndTabMenu > li:nth-child(6) > a").click()
    time.sleep(5)
    print(f"{i}진행중")
    for k in range(0,2):
        for j in range(1,11):
            movie_list.append(title[i])
            try:
                movie_title_list.append(driver.find_element(By.CSS_SELECTOR, f"#reviewTab > div > div > ul > li:nth-child({j}) > a > strong").text)
                movie_text_list.append(driver.find_element(By.CSS_SELECTOR, f"#reviewTab > div > div > ul > li:nth-child({j}) > p > a").text)
            except:
                continue
            time.sleep(2)
        driver.find_element(By.CSS_SELECTOR,"#pagerTagAnchor2").click()
    time.sleep(5)
```

```python
#title : #reviewTab > div > div > ul > li:nth-child(1) > a > strong
# text : #reviewTab > div > div > ul > li:nth-child(1) > p > a
for j in range(0,2):
    for i in range(1,11):
        movie_title_list.append(driver.find_element(By.CSS_SELECTOR, f"#reviewTab > div > div > ul > li:nth-child({i}) > a > strong").text)
        movie_text_list.append(driver.find_element(By.CSS_SELECTOR, f"#reviewTab > div > div > ul > li:nth-child({i}) > p > a").text)
        time.sleep(2)
        
    driver.find_element(By.CSS_SELECTOR,"#pagerTagAnchor2").click()
    
```
